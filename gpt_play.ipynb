```python
# and train generator on this text data
my_run_name = 'ign_345M_5000_epochs'
model_name = "345M"
if not os.path.isdir(os.path.join('models', model_name)):
	print(f"Downloading {model_name} model...")
	gpt2.download_gpt2(model_name=model_name, 
                       model_dir='models')   # model is saved into current directory under /models/124M/

file_name = 'gpt_language_training_data/ign_cleaned_reviews_01.txt'

sess = gpt2.start_tf_sess()
gpt2.finetune(sess,
              file_name,
              model_name=model_name,
              run_name=my_run_name,
              steps=5000)   # steps is max number of training steps
```

```python
# load model here
my_run_name = 'ign_345_try01'
model_name = "345M"
sess = gpt2.start_tf_sess()
gpt2.load_gpt2(sess, 
               run_name=my_run_name, 
               model_name=model_name)
```

```python
# have to select model here
prefix = 'Batman: Arkham City Review'
length = 1023
nsamples=6
temperature=1.0
batch_size=6
sample_dir = 'samples'
dir_path = 'samples/{}'.format(my_run_name)
if not os.path.isdir(dir_path):
    os.mkdir(dir_path)
write_path = '{}/sample.txt'.format(dir_path)
gpt2.generate(sess, 
              run_name=my_run_name, 
              model_name=model_name, 
              prefix=prefix, 
              temperature=temperature, 
              length=length, 
              nsamples=nsamples, 
              destination_path=write_path, 
              batch_size=batch_size) 
```

```python
def create_and_write_samples(prefix, length, nsamples, 
                             temperature, batch_size, 
                             sample_dir, 
                             run_name, iterator):
    
    dir_path = 'samples/{}'.format(run_name)
    if not os.path.isdir(dir_path):
        os.mkdir(dir_path)
    write_path = '{}/sample{}.txt'.format(dir_path, iterator)
    gpt2.generate(sess, 
                  run_name=run_name, 
                  model_name=model_name, 
                  prefix=prefix, 
                  temperature=temperature, 
                  length=length, 
                  nsamples=nsamples, 
                  destination_path=write_path, 
                  batch_size=batch_size) 
    
    return
```
